## Le LaTeX

---

L'intérêt particulièrement interressant dans ces cellules mardown est de pouvoir y méler des zones de langage $\LaTeX$ afin d'écrire les mathématique de façon simplifier et sans se soucier des problèmes de mise en forme car c'est l'interpréteur qui s'en occupera !
Pour ouvrir une zone $\LaTeX$ il suffit de placer un dollars (\$) de part et d'autre de ce que l'on souhaite insérer !

Sans $\LaTeX$ : \
x \
Avec $\LaTeX$ : \
$ x $

Si on ne souhaite pas inclure la zone $\LaTeX$ dans le texte, il est possible de créer une zone $\LaTeX$ centrée et isolée en plaçant cette fois-ci deux dollars de part et d'autre de ce que l'on souhaite insérer !

Avec 1 \$ : \
$ x $ \
Avec 2 \$ : 
$$ x $$

L'intérêt du $\LaTeX$ est d'écrire de belle formules mathématiques ! Avec les lettres grecques incluses, mais aussi tout un ensemble de fonction permettant de faire des fractions ou bien des limites...

$$\alpha, \beta, \gamma, \sigma, \omega $$

$$\frac{1}{x^2} = x^{-2}$$

$$ \underset{x \rightarrow + \infty}{\lim} \sqrt{x} = + \infty $$
