### Importation des librairies :

```python
import numpy as np
import matplotlib.pyplot as plt 
# On donne aux librairies des alias pour faciliter leur utilisation 
# A chaque fois que vous voudrez les utiliser il suffira d'utiliser l'alias suivie d'un point pour préciser que vous voulez utiliser les modules/ function de la librairie
```

### Découverte de Numpy : Les premiers Array 

```python
# Pour creer un array on utilise Numpy.array ou ici np.array avec np comme alias pour numpy
array1 = np.array([1,2,3,4,5]) # L'interet de numpy etant contrairement aux list (vu précédement) de permettre de faire des calculs numériques
list1 = [1,2,3,4,5] 
```

```python
print(f"Résultat du produit de 3 et d'un Array :{array1 * 3},\net ici il s'agit le resultat mais pour une list {list1 * 3}")
```

```python
# Lorsque l'on manipule des Array il est important de toujours vérifier leur taille 
# Deux Array de taille differente que l'on voudrait transformer ensemble pourrait mener a une erreur dite de Shape
array1.shape
np.shape(array1)
# On peut verifier la shape de deux manieres soit avec .shape si on sait qu'il s'agit d'un Array à 100%
# Sinon il vaut mieux dans le doute utiliser np.shape() qui lui s'applique à toutes les variables 
```

```python
# Numpy permet de créer des Array multidimensionels comme on peut le voir ici. 
# Array 2 est l'equivalent d'un tableau et posséde 2 dimensions.
array2 = np.array([[1,2,3], [3,2,1]])
print(array2)
```

```python
np.shape(array2) #np.shape() nous donne ici la taille de l'array2 en indiquant TOUJOURS ligne * colonne ( et si il y a d'autre dimensions la profondeur …)
```

```python
# De la même maniere que pour les list, il est possible d'indexer les Array et ainsi de ne selectionner que certaines portions 
# Néanmoins de part leur caractére multidimensionel il est important de preter attention à la dimensions que vous souhaiter indexer
# Tout comme la shape vous indexer tout d'abord sur les lignes, puis les colonnes …
print(f"La première ligne est : {array2[0]}, la seconde est : {array2[1]}")
```

```python
print(f"La première colonne est : {array2[:,0]}") # Ici je precise avec les ":" que je prend toutes les lignes mais avec le "0" que je ne veux que la premiere colonne 
```

```python
print(array2[:,2]) # Et ici je prend la troisième colonne 
```

```python
# Voici precisement un exemple d'Array avec une dimension encore plus grande 
array3 = np.array([[[1,2],[3,4]], [[5,6],[7,8]]])
print(array3)
```

```python
array3.shape
```

```python
# Numpy permet de creer directement des Array des grandes dimensions avec des valeurs.
# Ici arange permet de creer un Array contenant les nombres entres deux bornes tous espacés d'un meme pas/step
arange1 = np.arange(0, 200, 0.1) 
print(arange1) # arange1 contient tous les nombres de 0 à 200 non inclus tous séparés d'un pas de 0.1 : donc 0, 0.1, 0.2, 0.3, 0.4 … 
```

```python
arange1.shape # La shape d'un arange est donc de end/step : ici 200/0.1 = 2000
```

```python
# Il existe un autre moyen de faire des array espacé egalement mais cette fois si directement meme si on ne connais pas le step.
# On précise à np.linspace le depart, la fin (incluse cette fois) et le nombre de points que l'on veut
# Linspace va venir directement subdiviser l'espace de 0 à 200 uniformement 
linspace1 = np.linspace(0, 199, 200) 

print(linspace1.shape) 
```

```python
print(linspace1) 
```

```python
# Numpy fournit également des fonctions mathématiques plus avancée pour faire des calculs comme sinus, cosinus, log, exp …
# Pour se faire on utilise np. la fonction mathematique que l'on souhaite utiliser. Par exemple np.sqrt() pour racine carré 
# L'interet etant que cela s'apllique directement à l'ensemble des valeurs d'un Array, rendant le processus rapide

# Si on reprend l'array du début array1, on peut connaitre la racine carré de tous ces nombres. 
sqrt1 = np.sqrt(array1)
print(f"Le carré de {array1} est {sqrt1} \nOn voit que la fonction np.sqrt() permet de renvoyer une varibale de type {type(sqrt1)} ce qui est vachement pratique.")
```

### Matplotlib et les représentations graphique : 

```python
# Matplotlib contient enormément de fonction pour faire des représentation graphique en tout genre. 
# Les plus classique et par analogie à Matlab ou R vous retrouvez : plot, hist, bar, boxplot …
```

```python
# Plot par exemple prend deux varibale x et y 
# Ici x = arange1 et y = np.sin(arange1)-> donnant un array de même taille que arrange1 avec le resultat de la fonction sinus appliquée sur le vecteur
plt.plot(arange1, np.sin(arange1)) 
```

```python
# Prenons un exemple plus simple a visualiser avec un array de plus petite taille 
exp1 = np.exp(array1) # exp contient le resulat de la fonction exponetielle aplliquée sur l'array1

# Lancement du graph ;
plt.plot(array1, exp1)
# Ajout d'élements graphiques ;
plt.xlabel("X")
plt.ylabel("$Y =\exp^{(X)} $") # Comme vous le voyez on peut y mettre des elements latex aussi dans les string/str
plt.xticks(array1)
plt.title("Représentation graphique de la fonction exponentielle")
plt.show() # Permet de forcer l'affichaque de la figure et surtout de retirer les petits messages de matplotlib, essayez sans vous verrez la différence

# Il existe une multitude de représentations graphiques et ajouts stylistiques 
# De même matplotlib n'est pas la seule librairies de représentation graphique, mais les autres sont moins évidente et moins connue que matplotlib

# La seule limite pour la suite est votre imagination
```
